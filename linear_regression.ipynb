```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_diabetes
```

```python
diabetes = load_diabetes()
```

```python
diabetes
```

```python
print(diabetes.DESCR)
```

```python
#problem statment >> To predict disease disease progression one year after baseline. given 10 baselne feature
```

```python
diabetes.data
```

```python
diabetes.target
```

```python
diabetes.feature_names
```

```python
data = pd.DataFrame(diabetes.data, columns= diabetes.feature_names)
```

```python
data.target
```

```python
data['target'] = diabetes.target
```

```python
data
```

```python
# EDA, DATACLEANING DATA PREPARATION, FEATURE ENGINEERING
```

```python
# DIVID THE DATA IN X AND Y, for simple linear regression we take only one feature
```

```python
x = data[["bmi"]]
y = data["target"]
```

```python
x
```

```python
y
```

```python
# split the data into train-test
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test, y_train, y_test =  train_test_split(x,y, test_size=0.2, random_state=1)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test
```

```python
#Scaling(optional) This data is already scaled
```

```python
from sklearn.linear_model import LinearRegression
#Thats how we write the first model
```

```python
model = LinearRegression() #we will do always instantiate the object object instance like that 
```

```python
model
```

```python
#For training the model humm is model mein
model.fit(x_train,y_train) #at backened it will use gradient descent to give optimal co_officient
```

```python
model.coef_
#coeff are optimal beta1 which means slope
```

```python
model.intercept_
```

```python
#now we will predict model has been train
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
#lets visual the result
plt.scatter(x_test, y_test, color = 'black', label = 'Actutal data')
plt.plot(x_test, y_pred, color = 'blue', linewidth = 3, label = "Linear regression line")
plt.xlabel("BMI")
plt.ylabel("One year progression-target")
plt.title("Linear Regression on diabetes data")
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test, y_pred)
```
